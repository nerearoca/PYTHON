## Funciones

````python
# para encapsular codigo, podemos definir funciones de esta manera

def saludame(tu_nombre, tu_edad):
  comentario_edad = ""  # declaramos un string vacio que se rellenará a continuación
  if tu_edad > 35:
    comentario_edad = "viejo de narices"
  else:
    comentario_edad = "demasiado joven"

  # esto es un template, rellena lo que hay entre llaves con una variable. requiere poner la "f" delante de las comillas
  # tipo el printf de C
  return f"Hola {tu_nombre}, eres {comentario_edad}"

# llamamos a la funcion asignando valores a parametro explicitamente. No hace falta que respetemos el orden de los parámetros si se hace asi
retorno = saludame(tu_nombre="joan", tu_edad=37)
print(retorno)

print()

# la llamamos asignando los parámetros en el orden esperado.
nombre = "pepe"
edad = 19
retorno = saludame(nombre, edad)
print(retorno)
````

````python
# una función puede tener parámetros opcionales o con valores por defecto

def func(param_obligatorio, param_opcional=None, param_con_default=3):
  # checkeamos si param_opcional tiene valor. Es lo mismo que hacer param_opcional is not None
  if param_opcional:
    return param_obligatorio + param_opcional + param_con_default
  else:
    return param_obligatorio + param_con_default

# podemos llamar de estas maneras
x = 10
y = 20
z = 30

# con valor para la opcional y la que tiene un valor por defecto
print(func(param_obligatorio=x, param_opcional=y, param_con_default=z))

# sin pasar la opcional pero con valor la que tiene por defecto
print(func(param_obligatorio=x, param_con_default=z))

# pasando solo la obligatoria
print(func(param_obligatorio=x))
````

## Clases

````python
# Una clase se define asi

class UnaClase:

  # metodo que recoje los parámetros de entrada de la clase, es opcional
  # los metodos de clase requieren el self como primer parametro
  def __init__(self, un_parametro):
    # guardamos el parametro como atributo de la clase
    self.un_parametro = un_parametro

  def una_funcion(self):
    # operamos sobre un atributo de la clase
    return self.un_parametro * 10


# instanciamos la clase. Necesitamos pasar los parámetros del init
instancia = UnaClase(un_parametro=30)

# llamamos a la funcion
retorno = instancia.una_funcion()
print(retorno)
````

Python tiene muchísimas librerias especializadas en diferentes tareas de procesado del lenguaje. El Natural Language Toolkit (NLTK) es una de las que históricamente se ha usado más. Así y todo hoy en día se ha quedado un poco obsoleta y la gente usa sobretodo Spacy.

````python
import nltk

# we download extra nltk resources
nltk.download("punkt")
nltk.download('averaged_perceptron_tagger')
````

````python
#EJECUTAR LA CELDA DE ARRIBA PRIMERO!

import nltk

text = "This is a sentence. This is another sentence. And now we have the third one"

sent_text = nltk.sent_tokenize(text)

# dividimos el texto en frases y iteramos sobre cada una
for sentence in sent_text:
    print("---> Frase")
    print(sentence)

    # dividimos la frase en palabras
    tokenized_text = nltk.word_tokenize(sentence)
    print("---> Palabras")
    print(tokenized_text)


    # calculamos las categorias gramaticales de cada palabra
    tagged = nltk.pos_tag(tokenized_text)
    print("---> Categorias gramaticales por palabra")
    print(tagged)

    print()
````

````python
# Vamos a probar Spacy
import spacy

# cargamos un modelo. Esto contiene todo lo necesario para hacer extracción de información de textos.
# hay modelos para todos los idiomas.
# la documentacion con todas las : https://spacy.io/usage/linguistic-features

nlp = spacy.load("en_core_web_md")
text = "Are you guys having fun?"

doc = nlp(text)
print(doc.text)
print()

for token in doc:
    # printamos la palabra, la categoria gramatical y la palabra en su forma base
    # el lemma pone las palabras en singular y deja los verbos en su forma base
    print(token.text, token.pos_, token.lemma_)
````

# ML Básico

````python
# definamos una clase Animal

class Animal:
    def __init__(self, altura: int, peso: int, tipo: str):
      self.altura = altura
      self.peso = peso
      self.tipo = tipo

    # devuelve una lista con la altura en la posicion 0 y el peso en la 1
    def get_features(self):
      return [self.altura, self.peso]

    # devuelve la etiqueta
    def get_label(self):
      return self.tipo
````

````python
# creemos un dataset con gatos y leones

leon_1 = Animal(altura=120, peso=170, tipo="leon")
leon_2 = Animal(altura=110, peso=180, tipo="leon")
leon_3 = Animal(altura=90, peso=140, tipo="leon")
leon_4 = Animal(altura=80, peso=130, tipo="leon")
leon_5 = Animal(altura=130, peso=190, tipo="leon")

gato_1 = Animal(altura=30, peso=4, tipo="gato")
gato_2 = Animal(altura=40, peso=6, tipo="gato")
gato_3 = Animal(altura=25, peso=3, tipo="gato")
gato_4 = Animal(altura=45, peso=9, tipo="gato")
gato_5 = Animal(altura=24, peso=7, tipo="gato")

# matriz donde tenemos 10 filas y 2 columnas
# cada fila es un animal. Cada columna es una feature
# en la columna 0 -> altura, en la columna 1 -> peso

features = [
              leon_1.get_features(), leon_2.get_features(), leon_3.get_features(), leon_4.get_features(), leon_5.get_features(),
              gato_1.get_features(), gato_2.get_features(), gato_3.get_features(), gato_4.get_features(), gato_5.get_features()
           ]

# lista donde en la posición "i" tenemos el tipo de animal. Si usamos "i" en features, tenemos sus features correspondientes
labels = [
              leon_1.get_label(), leon_2.get_label(), leon_3.get_label(), leon_4.get_label(), leon_5.get_label(),
              gato_1.get_label(), gato_2.get_label(), gato_3.get_label(), gato_4.get_label(), gato_5.get_label()
           ]

# mismos indices contienen las features y el label correspondiente en features y labels
print(features[0], labels[0])
````

````python
# Entrenemos ahora un clasificador con este dataset
# Un clasificador extrae patrones estadísticos de los datos de entrenamiento.
# Podemos testearlo posteriormente con datos que no ha visto antes

from sklearn.svm import SVC

clf = SVC(probability=True)
trained_classifier = clf.fit(features, labels)
````

````python
# Testeamos el modelo
# Editad los valores para ver como evolucionan las predicciones
altura = 100
peso = 50


test_instance = [[altura, peso]]
prediction = trained_classifier.predict(test_instance)[0]
print(f"Predicción: {prediction}")
print()

probabilities = trained_classifier.predict_proba(test_instance)[0]
class_names = trained_classifier.classes_

print("Probabilidades por tipo")
for class_name, class_probability in zip(class_names, probabilities):
  print(class_name, class_probability)
````

````python
import spacy
nlp = spacy.load("en_core_web_md")
````

````python
# ejecutad la celda de arriba!

text1 = "Factory"
text2 = "Industry"

token_info1 = nlp(text1)[0]
token_info2 = nlp(text2)[0]

# Descomentar para ver los vectores que ha sacado para cada palabra. Como vereis, son basicamente vectores gigantes con numeros que a priori poco nos dicen
# print(token_info1.vector)
# print(token_info2.vector)

# Observad la similitud semantica que saca entre palabras.
# Cambiad text1 y text2 para ver como se comporta esta metrica
print(f"Similitud entre las palabras {text1}, {text2}")
similarity = token_info1.similarity(token_info2)
print(similarity)
````

````python
# Vamos a hacer un grafico para ver como las palabras quedan representadas en un espacio semantico 2D
# Rellenad la lista list_of_words de palabras y observad como hay grupos medianamente interpretables

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# this line makes any plots display in the notebook
%matplotlib inline

def get_word_vectors(words):
    # converts a list of words into their word vectors
    return [nlp(word).vector for word in words]


def visualize(words):
    pca = PCA(n_components=2)
    pca.fit(get_word_vectors(words))
    word_vecs_2d = pca.transform(get_word_vectors(words))
    plt.figure(figsize=(10,5))
    plt.scatter(word_vecs_2d[:,0], word_vecs_2d[:,1])

    # for each word and coordinate pair: draw the text on the plot
    for word, coord in zip(words, word_vecs_2d):
        x, y = coord
        plt.text(x, y, word, size= 15)

    # show the plot
    plt.show()

list_of_words = ['dolphin', 'salmon', 'data', 'fish' , 'lion', 'her', 'she', 'travel','holiday',  'him', 'info', 'file']

visualize(list_of_words)

````

````python
###################
#### SOLUCION!! ###
###################
"""
LOW QUALITY ASCII ART
       ,___          .-;'
       `"-.`\_...._/`.`
    ,      \        /
 .-' ',    / ()   ()\
`'._   \  /()    .  (|
    > .' ;,     -'-  /
   / <   |;,     __.;
   '-.'-.|  , \    , \
      `>.|;, \_)    \_)
       `-;     ,    /
          \    /   <
           '. <`'-,_)
           '._)

"""

def similar_pairs(doc):
    processed_pairs = set()
    output = []

    for word1 in doc:
        for word2 in doc:
            if (word1.text != word2.text) and ((word1.text, word2.text)) not in processed_pairs:
                similarity = word1.similarity(word2)
                tuple_similarity = (word1.text, word2.text, similarity)
                output.append(tuple_similarity)

                # metemos A,B y B,A porque es lo mismo
                processed_pairs.add((word1.text, word2.text))
                processed_pairs.add((word2.text, word1.text))

    # ordenamos la lista por el tercer componente de la tupla
    ordered_output = sorted(output, key=lambda tup: tup[2], reverse=True)
    return ordered_output


text = "In 2019, Emma Watson attended the United Nations General Assembly in New York City, where she met with Bill Gates and Malala Yousafzai to discuss the impact of the HeForShe campaign. Meanwhile, Elon Musk was in California, preparing for the launch of SpaceX’s Falcon Heavy rocket from Cape Canaveral, Florida. At the same time, J.K. Rowling was in Edinburgh, Scotland, working on her next novel, while the Louvre in Paris hosted a special exhibition featuring works by Leonardo da Vinci, including the famous “Mona Lisa.”"
doc = nlp(text)
pairs = similar_pairs(doc)

# print top 30
for word1, word2, similarity in pairs[:30]:
  print(word1, word2, similarity)
````

````python
# vamos a descargarnos un dataset de reviews de YELP
# este dataset contiene un texto y 0 si es negativa la review o 1 si es positiva.
# entrenemos un modelo para poder hacer sentiment analysis sobre textos no vistos antes

from datasets import load_dataset
dataset = load_dataset('yelp_polarity', split='train')


# exploremos un par de elementos del dataset
print(dataset[0]["text"])
print(dataset[0]["label"])
print()
print(dataset[1]["text"])
print(dataset[1]["label"])
print()
print(dataset[2]["text"])
print(dataset[2]["label"])
````

````python
# Ejemplo completo de uso de embeddings como representacion en features de textos

import numpy as np

# funcion para calcular el embedding medio de un texto
# se saca el embedding de cada palabra y se hace un average dimension a dimension
def calcular_embedding_spacy(nlp, text):
    doc = nlp(text)
    vectors = []
    for token in doc:
        vector = token.vector
        vectors.append(vector)

    # hace la media dimension a dimension, es decir, en la posicion 0 habra la media de todas las posiciones 0
    # asi para todas
    avg_vector = np.mean(np.array(vectors), axis=0)
    return avg_vector


import spacy
nlp = spacy.load("en_core_web_md")

# seleccionemos N textos. Los usaremos para entrenar un modelo de detección de sentimiento: 0 -> negativo. 1 -> positivo
N = 1500
subset_train = dataset[0:N]

training_vectors = []
labels_train = []
# para cada texto del training set, calculemos su representación vectorial mediante spacy
for i in range(0, N):
    text = subset_train["text"][i]
    label = subset_train["label"][i]
    vector = calcular_embedding_spacy(nlp, text)
    training_vectors.append(vector)
    labels_train.append(label)


# ahora seleccionemos otro subset de test
M = 100
subset_test = dataset[N+1:N+1+M]

test_vectors = []
labels_test = []

# para cada texto del test set, calculemos su representación vectorial mediante spacy
for i in range(0, M):
    text = subset_test["text"][i]
    label = subset_test["label"][i]
    vector = calcular_embedding_spacy(nlp, text)
    test_vectors.append(vector)
    labels_test.append(label)

# entrenemos el modelo
from sklearn.svm import SVC
clf = SVC(probability=True)
trained_classifier = clf.fit(training_vectors, labels_train)
````

````python
# testead el modelo, introducid un texto en ingles. Cambiadlo y mirad como varia la clasificación
test_text = "This movie is amazing, I am very happy"

# calculamos el embedding del texto de test
vector_test = calcular_embedding_spacy(nlp, test_text)
prediction = trained_classifier.predict([vector_test])[0]
print(f"Predicción: {prediction}")
print()
````

````python
# ahora veamos como de bien funciona el modelo en el test set
predictions = clf.predict(test_vectors)
from sklearn.metrics import classification_report
print(classification_report(labels_test, predictions))

# MINI EJERCICIO!!!
# VARIAD N y M de la celda de entrenamiento
# Mirad como mejora la precision del modelo cuantos mas datos tiene de entreno
````

````python
# Hagamos lo mismo, pero con vectores mas potentes. Tardara un poco mas

from sentence_transformers import SentenceTransformer

def calcular_embedding_transformers(model, text):
    embeddings = model.encode(text)
    return embeddings

# Estos embeddings son multilingues y a nivel de texto!!!!
model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')

# seleccionemos N textos. Los usaremos para entrenar un modelo de detección de sentimiento: 0 -> negativo. 1 -> positivo
N = 500
subset_train = dataset[0:N]

training_vectors = []
labels_train = []
# para cada texto del training set, calculemos su representación vectorial mediante transformers
for i in range(0, N):
    text = subset_train["text"][i]
    label = subset_train["label"][i]
    vector = calcular_embedding_transformers(model, text)
    training_vectors.append(vector)
    labels_train.append(label)


# ahora seleccionemos otro subset de test
M = 100
subset_test = dataset[N+1:N+1+M]

test_vectors = []
labels_test = []

# para cada texto del test set, calculemos su representación vectorial mediante transformers
for i in range(0, M):
    text = subset_test["text"][i]
    label = subset_test["label"][i]
    vector = calcular_embedding_transformers(model, text)
    test_vectors.append(vector)
    labels_test.append(label)


# entrenemos el modelo
from sklearn.svm import SVC
clf_transformers = SVC(probability=True)
clf_transformers.fit(training_vectors, labels_train)
````

````python
# testead el modelo, introducid un texto en el idioma que querais. Cambiadlo y mirad como varia la clasificación
test_text = "Aquesta pelicula ha estat una sorpresa molt agradable, crec que es una de les millors pelis que he vist mai"

# calculamos el embedding del texto de test
vector_test = calcular_embedding_transformers(model, test_text)
prediction = clf_transformers.predict([vector_test])[0]
print(f"Predicción: {prediction}")
print()
````

````python
# comparemos este modelo ya entrenado con los otros clasificadores que hemos creado

def get_predictions_pretrained(pretrained_model, texts):
    normalized_predictions = []
    kwargs = {'padding':True,'truncation':True,'max_length':512}
    predictions = pretrained_model(texts, **kwargs)
    for prediction in predictions:
        # este modelo da predicciones diferentes, decidimos que si es 1 star o 2 star es negativo
        # si es otro valor, lo consideraremos positivo
        normalized_prediction = 1
        if prediction["label"] in ["1 star", "2 stars"]:
            normalized_prediction = 0

        normalized_predictions.append(normalized_prediction)

    return normalized_predictions


N = 500
M = 100
subset_test = dataset[N+1:N+1+M]

test_texts = []
labels_test = []

# para cada texto del test set, obtenemos el texto y su polaridad
for i in range(0, M):
    text = subset_test["text"][i]
    label = subset_test["label"][i]
    test_texts.append(text)
    labels_test.append(label)

predictions = get_predictions_pretrained(pretrained_model, test_texts)
from sklearn.metrics import classification_report
print(classification_report(labels_test, predictions))

# No esta mal! Un modelo que ya nos lo dan hecho tiene una performance bastante buena!
````

````python
from openai import OpenAI
from google.colab import userdata
import os
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')

def send_prompt(client, prompt, system_prompt=None):
    if not system_prompt:
        messages = [{"role": "user", "content": prompt}]
    else:
        messages = [
                      {"role": "system", "content": system_prompt},
                      {"role": "user", "content": prompt}
                   ]

    response_gpt = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages
    )

    answer = response_gpt.choices[0].message.content
    return answer


client = OpenAI()
text = "Cuentame un chiste mi amigo"
response = send_prompt(client, text)
print(response)
````
